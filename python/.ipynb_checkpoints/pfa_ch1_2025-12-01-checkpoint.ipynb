```python
#ダックタイピング ・・・ オブジェクトの型は重視しない。実行できればOK 
def isiterable(obj):
    try:
        iter(obj)
        return True
    except TypeError: #not iterable
        return False

isiterable("a string")
isiterable([1, 2, 3])
isiterable(5)

# iterは繰り返し可能オブジェクト(=イテレータ)への変換関数。try:iter()
numbers = [1, 2, 3]
it = iter(numbers)

print(next(it))
print(next(it))
```

```python
#nullはNone
b = 4
b is not None

#関数の引数としてよく使われる
def add_and_maybe_multiply(a, b, c=None):
    result = a + b
    if c is not None:
        result *= c
    return result

print(add_and_maybe_multiply(3, 3))
print(add_and_maybe_multiply(3,3,3))
```

```python
#問題1-1
scores = [72, 85, 90, 55, 100, 67,88,92,45,79]
ave = 0
top = 0
bottum = 100
for i in range(len(scores) - 1):
    if scores[i] > top:
        top = scores[i]
    if scores[i] < bottum:
        bottum = scores[i]
    ave += scores[i]/(len(scores))

print(ave, top, bottum)

list = []
#問題1-2
for i in range(len(scores) -1 ):
    if scores[i] < 70:
        list.append(scores[i])
    else:
        continue 

print(list, len(list))

rate = 0
#問題1-3
for i in range(len(scores) -1):
    if scores[i] >= 80:
        rate += 1
    else:
        continue
print( (rate/len(scores)) * 100)
        
```

```python
#問題２
students = [
    {"name": "Alice", "math": 85, "bio": 92},
    {"name": "Bob",   "math": 70, "bio": 65},
    {"name": "Carol", "math": 95, "bio": 88},
    {"name": "Dave",  "math": 60, "bio": 73},
]

#1
for a in students:
    a["average"] = (a["math"] + a["bio"])/2

print(students)

ave = 0
for a in students:
    ave += a["average"]/(len(students))
print(ave)

list = []
for a in students:
    if a["average"] > ave:
        list.append(a["name"])
print(list)
```

```python
#問題3
logs = [
    "/home",
    "/about",
    "/home",
    "/products",
    "/home",
    "/products",
    "/contact",
    "/home",
    "/about",
]

paths = {}
for log in logs:
    count = 0
    for path in paths:
        if log == path.keys():
            paths[path.keys()] += 1
        elif count == len(paths):
            paths[log] += 1
        else:
            count += 1

top = { "path": 0 }
for path in paths:
    if path.values() > top.values():
        top = {path.keys() : path.values()}

sorted = []
for i in range(len(paths) -1):
    max = 0
    for path in paths:
        if path.values() > max:
            key = path.keys
    sorted.append((key, paths[key]))
    paths[key] = 0

        
```
